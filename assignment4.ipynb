```python
# allow output from every line
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

import pandas as pd
import numpy as np
```

```python
# Step 2-3:  Choose teams, get CSVs, make player name the index for each table
bears_2023_roster = pd.read_csv("data/Bears2023Roster.csv", index_col='Player')
bears_2023_roster
lions_2023_roster = pd.read_csv("data/Lions2023Roster.csv", index_col='Player')
lions_2023_roster
packers_2023_roster = pd.read_csv("data/Packers2023Roster.csv", index_col='Player')
packers_2023_roster
vikings_2023_roster = pd.read_csv("data/Vikings2023Roster.csv", index_col='Player')
vikings_2023_roster
```

```python
# Step 4:  Concatenate 4 tables
nfc_north_2023_rosters = pd.concat([bears_2023_roster,lions_2023_roster,packers_2023_roster,vikings_2023_roster])
nfc_north_2023_rosters
```

```python
# Step 5: Write new CSV
nfc_north_2023_rosters.to_csv("data/NFCnorth2023rosters.csv", na_rep="NAN")
```
